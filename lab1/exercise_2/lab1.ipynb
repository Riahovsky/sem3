```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd





OY = []
OX = []

try :
    with open('Dice.txt', 'r') as openedFile :
        for line in openedFile :
            tab = line.split()
            OY.append(float(tab[0]))
except IOError :
    print("IOError!")

fig = plt.figure()
plt.title('Dice')
plt.xlabel('number')
plt.ylabel('probability')

width = .50
ind = np.arange(len(OY))
plt.bar(ind, OY, width=width)

plt.savefig("Dice.pdf")

#######################################################################
OY = []
OX = []

try :
    with open('BonusDice.txt', 'r') as openedFile :
        for line in openedFile :
            tab = line.split()
            OY.append(float(tab[0]))
except IOError :
    print("IOError!")

fig = plt.figure()
plt.title('BonusDice')
plt.xlabel('number')
plt.ylabel('probability')

width = .50
ind = np.arange(len(OY))
plt.bar(ind, OY, width=width)


plt.savefig("BonusDice.pdf")

###################################################################
OY = []
OX = []

try :
    with open('PenaltyDice.txt', 'r') as openedFile :
        for line in openedFile :
            tab = line.split()
            OY.append(float(tab[0]))
except IOError :
    print("IOError!")

fig = plt.figure()
plt.title('PenaltyDice')
plt.xlabel('number')
plt.ylabel('probability')

width = .50
ind = np.arange(len(OY))
plt.bar(ind, OY, width=width)


plt.savefig("PenaltyDice.pdf")

################################################################
OY = []
OX = []

try :
    with open('ThreeDicePool.txt', 'r') as openedFile :
        for line in openedFile :
            tab = line.split()
            OY.append(float(tab[0]))
except IOError :
    print("IOError!")

fig = plt.figure()
plt.title('ThreeDicePool')
plt.xlabel('number')
plt.ylabel('probability')

width = .50
ind = np.arange(len(OY))
plt.bar(ind, OY, width=width)


plt.savefig("ThreeDicePool.pdf")
```
