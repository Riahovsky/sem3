```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd





OY = []
OX = []

try :
    with open('DoubleDice.txt', 'r') as openedFile :
        for line in openedFile :
            tab = line.split()
            OY.append(float(tab[0]))
except IOError :
    print("IOError!")

fig = plt.figure()
plt.title('DoubleDice')
plt.xlabel('number')
plt.ylabel('probability')

width = .50
ind = np.arange(len(OY))
plt.bar(ind, OY, width=width)

plt.savefig("DoubleDice.pdf")
```
